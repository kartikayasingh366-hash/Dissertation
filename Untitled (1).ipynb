```python
import re
import html
import unicodedata
from datetime import datetime

import numpy as np
import pandas as pd
from tqdm import tqdm

# Sentiment
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
from textblob import TextBlob

# BERT/Transformer sentiment
from transformers import pipeline


# -----------------------------
# Config: your file paths
# -----------------------------
GOOGLE_REVIEWS_PATH = "Google_reviews (1) (1).xlsx"
X_SCRAPED_PATH = "x_scraped_dataset.xlsx"

OUT_CLEAN_PATH = "combined_clean.csv"
OUT_SCORED_PATH = "combined_with_sentiment.csv"


# -----------------------------
# Text cleaning helpers
# -----------------------------
URL_RE = re.compile(r"(https?://\S+|www\.\S+)")
MENTION_RE = re.compile(r"@\w+")
HASHTAG_RE = re.compile(r"#\w+")
MULTISPACE_RE = re.compile(r"\s+")
NON_TEXT_RE = re.compile(r"[^0-9A-Za-z\s\.\,\!\?\:\;\'\"\-\(\)]")

def normalize_unicode(text: str) -> str:
    # Normalize weird unicode forms and strip accents safely
    text = unicodedata.normalize("NFKC", text)
    return text

def basic_clean_text(text: str) -> str:
    """
    A practical cleaning function aligned with your proposal’s preprocessing stage:
    tokenization/normalization, removing noise (URLs, mentions), and standardizing whitespace.
    """
    if text is None or (isinstance(text, float) and np.isnan(text)):
        return ""

    text = str(text)
    text = html.unescape(text)
    text = normalize_unicode(text)

    # Remove URLs, @mentions; keep hashtag words but remove the '#'
    text = URL_RE.sub(" ", text)
    text = MENTION_RE.sub(" ", text)
    text = text.replace("\n", " ").replace("\r", " ").replace("\t", " ")

    # Convert hashtags like #Tesco -> Tesco
    text = re.sub(r"#(\w+)", r"\1", text)

    # Remove most non-text symbols but keep basic punctuation
    text = NON_TEXT_RE.sub(" ", text)

    # Collapse whitespace
    text = MULTISPACE_RE.sub(" ", text).strip()

    return text


# -----------------------------
# Loaders for your datasets
# -----------------------------
def load_google_reviews(path: str) -> pd.DataFrame:
    df = pd.read_excel(path)

    # Your columns (observed):
    # ['Brand Name','Market Type','Country','Brand Origin','Date','Reviwer_Name','Rating','Review_description']
    rename_map = {
        "Brand Name": "retailer",
        "Market Type": "market_type",
        "Country": "country",
        "Brand Origin": "brand_origin",
        "Date": "date",
        "Reviwer_Name": "author_id",
        "Rating": "rating",
        "Review_description": "text",
    }
    df = df.rename(columns=rename_map)

    df["source_platform"] = "Google_Reviews"
    df["engagement_count"] = np.nan
    df["post_id"] = np.nan

    # parse date
    df["date"] = pd.to_datetime(df["date"], errors="coerce")

    # keep only relevant columns
    keep = [
        "source_platform", "post_id", "date", "author_id",
        "country", "brand_origin", "retailer", "market_type",
        "rating", "engagement_count", "text"
    ]
    return df[keep]


def load_x_scraped(path: str) -> pd.DataFrame:
    df = pd.read_excel(path)

    # Your columns (observed):
    # ['Country','Brand_Origin','Retailer','Source_Platform','Post_ID','Date',
    #  'Author_ID_Anon','Engagement_Count','Post_Text','Sentiment_VADER']
    rename_map = {
        "Source_Platform": "source_platform",
        "Post_ID": "post_id",
        "Date": "date",
        "Author_ID_Anon": "author_id",
        "Country": "country",
        "Brand_Origin": "brand_origin",
        "Retailer": "retailer",
        "Engagement_Count": "engagement_count",
        "Post_Text": "text",
        "Sentiment_VADER": "sentiment_vader_existing",
    }
    df = df.rename(columns=rename_map)

    df["rating"] = np.nan
    df["market_type"] = np.nan

    df["date"] = pd.to_datetime(df["date"], errors="coerce")

    keep = [
        "source_platform", "post_id", "date", "author_id",
        "country", "brand_origin", "retailer", "market_type",
        "rating", "engagement_count", "text", "sentiment_vader_existing"
    ]
    return df[keep]


# -----------------------------
# Sentiment scorers
# -----------------------------
def add_vader(df: pd.DataFrame, text_col="text_clean") -> pd.DataFrame:
    analyzer = SentimentIntensityAnalyzer()
    scores = df[text_col].fillna("").apply(lambda t: analyzer.polarity_scores(t))
    df["vader_neg"] = scores.apply(lambda d: d["neg"])
    df["vader_neu"] = scores.apply(lambda d: d["neu"])
    df["vader_pos"] = scores.apply(lambda d: d["pos"])
    df["vader_compound"] = scores.apply(lambda d: d["compound"])
    return df

def add_textblob(df: pd.DataFrame, text_col="text_clean") -> pd.DataFrame:
    def tb_polarity(t: str) -> float:
        if not t:
            return 0.0
        return float(TextBlob(t).sentiment.polarity)

    def tb_subjectivity(t: str) -> float:
        if not t:
            return 0.0
        return float(TextBlob(t).sentiment.subjectivity)

    df["textblob_polarity"] = df[text_col].fillna("").apply(tb_polarity)
    df["textblob_subjectivity"] = df[text_col].fillna("").apply(tb_subjectivity)
    return df

def add_bert_sentiment(df: pd.DataFrame, text_col="text_clean") -> pd.DataFrame:
    """
    Proposal mentions fine-tuned BERT; if you don't have labels to fine-tune yet,
    this uses a strong pretrained sentiment model as a drop-in baseline.
    """
    # Good general baseline; works well on social text too.
    clf = pipeline("sentiment-analysis", model="cardiffnlp/twitter-roberta-base-sentiment-latest")

    labels = []
    confs = []

    for t in tqdm(df[text_col].fillna("").tolist(), desc="BERT sentiment"):
        if not t.strip():
            labels.append("neutral")
            confs.append(0.0)
            continue
        out = clf(t[:512])[0]  # truncate long text
        labels.append(out["label"])
        confs.append(float(out["score"]))

    df["bert_label"] = labels
    df["bert_confidence"] = confs

    # Optional: map label -> numeric score for easier stats
    # cardiffnlp model labels are usually: negative / neutral / positive
    mapping = {"negative": -1.0, "neutral": 0.0, "positive": 1.0}
    df["bert_score"] = df["bert_label"].str.lower().map(mapping).fillna(0.0)
    return df


# -----------------------------
# Main pipeline
# -----------------------------
def main():
    # 1) Load
    g = load_google_reviews(GOOGLE_REVIEWS_PATH)
    x = load_x_scraped(X_SCRAPED_PATH)

    # 2) Combine
    df = pd.concat([g, x], ignore_index=True)

    # 3) Clean text
    df["text_clean"] = df["text"].apply(basic_clean_text)

    # Drop empty text rows (usually useless for sentiment)
    df = df[df["text_clean"].str.len() > 0].copy()

    # 4) Save cleaned-only
    df.to_csv(OUT_CLEAN_PATH, index=False)

    # 5) Sentiment scoring (proposal stack: VADER + TextBlob + BERT)
    df = add_vader(df, "text_clean")
    df = add_textblob(df, "text_clean")
    df = add_bert_sentiment(df, "text_clean")

    # 6) Save scored dataset
    df.to_csv(OUT_SCORED_PATH, index=False)

    print("Saved:", OUT_CLEAN_PATH)
    print("Saved:", OUT_SCORED_PATH)
    print("Rows:", len(df))


if __name__ == "__main__":
    main()
```

```python
# analysis_steps.py

import os
import re
import numpy as np
import pandas as pd

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation

from scipy.stats import ttest_ind, levene
import statsmodels.formula.api as smf


# -----------------------------
# Paths (use your existing outputs)
# -----------------------------
DATA_PATH = "combined_with_sentiment.csv"  # you uploaded this
OUT_DIR = "outputs_step2_step3"
os.makedirs(OUT_DIR, exist_ok=True)

RQ_OUT = os.path.join(OUT_DIR, "research_questions_and_hypotheses.md")
LDA_TOPIC_OUT = os.path.join(OUT_DIR, "lda_topics.csv")
LDA_DOC_TOPIC_OUT = os.path.join(OUT_DIR, "lda_doc_topic_distribution.csv")
STATS_OUT = os.path.join(OUT_DIR, "local_global_stats_tests.csv")
REG_OUT = os.path.join(OUT_DIR, "regression_summary.txt")


# -----------------------------
# Helpers
# -----------------------------
def pick_sentiment_column(df: pd.DataFrame) -> str:
    """
    Prefer fine-tuned BERT score if present, otherwise fallback.
    """
    for c in ["ft_bert_score", "bert_score", "vader_compound", "textblob_polarity"]:
        if c in df.columns:
            return c
    raise ValueError("No supported sentiment column found. Expected one of: "
                     "ft_bert_score, bert_score, vader_compound, textblob_polarity")

def normalize_brand_origin(df: pd.DataFrame) -> pd.DataFrame:
    """
    Force brand_origin into Local/Global if possible.
    Accepts values like: local/global, domestic/foreign, etc.
    """
    if "brand_origin" not in df.columns:
        raise ValueError("brand_origin column not found in dataset.")

    s = df["brand_origin"].astype(str).str.strip().str.lower()

    # common mappings
    local_vals = {"local", "domestic", "home", "national"}
    global_vals = {"global", "international", "foreign", "multinational"}

    def map_origin(x: str):
        if x in local_vals:
            return "Local"
        if x in global_vals:
            return "Global"
        # fallback: try substring match
        if "local" in x or "domestic" in x:
            return "Local"
        if "global" in x or "foreign" in x or "international" in x:
            return "Global"
        return np.nan

    df["brand_origin_norm"] = s.apply(map_origin)
    return df

def write_rqs_and_hypotheses(path: str):
    """
    Generates a clean, dissertation-ready RQ/H section matching your proposal.
    Edit wording if needed, but it’s immediately usable.
    """
    content = """# Research Questions & Hypotheses (to paste into dissertation)

## Research Questions (RQs)

**RQ1.** Do consumers express different **sentiment polarity** toward **Local vs Global** retailers?  
**RQ2.** Does **sentiment variance** (emotional intensity/dispersion) differ between Local vs Global retailers?  
**RQ3.** Does the Local vs Global difference in sentiment **vary by country** (country as a proxy for cultural context)?  
**RQ4.** What are the main **themes/topics** in consumer discourse, and do topic prevalence differ by Local vs Global retailers and by platform (Google Reviews vs X)?

## Hypotheses (H)

**H1 (Mean polarity).** Average sentiment polarity differs between **Local** and **Global** retailers.  
- Alternative directional version (use only if you can justify with literature): Local retailers receive higher mean sentiment.

**H2 (Variance / emotionality).** Sentiment variance differs between Local and Global retailers.  
- Alternative directional version: Local retailers elicit **higher variance** (more emotionally charged sentiment).

**H3 (Country moderation).** The relationship between brand origin (Local vs Global) and sentiment is **moderated by country**  
(i.e., an interaction effect: origin × country).

**H4 (Topics).** The distribution of discussion topics differs by brand origin and platform.
"""
    with open(path, "w", encoding="utf-8") as f:
        f.write(content)


def run_lda(
    df: pd.DataFrame,
    text_col: str,
    n_topics: int = 8,
    max_features: int = 6000,
    min_df: int = 10,
    max_df: float = 0.90,
    random_state: int = 42,
):
    """
    LDA using CountVectorizer + LatentDirichletAllocation.
    Returns: lda_model, vectorizer, doc_topic_matrix (n_docs x n_topics), feature_names
    """
    # Basic English stopwords; you can extend with domain-specific words if needed.
    vectorizer = CountVectorizer(
        stop_words="english",
        max_df=max_df,
        min_df=min_df,
        max_features=max_features,
        ngram_range=(1, 2),
    )

    X = vectorizer.fit_transform(df[text_col].fillna("").astype(str).tolist())

    lda = LatentDirichletAllocation(
        n_components=n_topics,
        learning_method="batch",
        random_state=random_state,
        max_iter=20,
        evaluate_every=-1,
    )
    doc_topic = lda.fit_transform(X)
    feature_names = vectorizer.get_feature_names_out()

    return lda, vectorizer, doc_topic, feature_names


def top_words_per_topic(lda, feature_names, top_n=12):
    topics = []
    for k, comp in enumerate(lda.components_):
        top_idx = np.argsort(comp)[::-1][:top_n]
        words = [feature_names[i] for i in top_idx]
        topics.append({"topic_id": k, "top_terms": ", ".join(words)})
    return pd.DataFrame(topics)


def stats_local_global(
    df: pd.DataFrame,
    sentiment_col: str,
    group_cols=("source_platform", "country"),
):
    """
    Per (platform, country), test:
      - Mean difference Local vs Global: Welch t-test
      - Variance difference: Levene test (Brown-Forsythe style by using median center)
    Outputs a table.
    """
    rows = []
    for keys, sub in df.groupby(list(group_cols)):
        sub = sub.dropna(subset=["brand_origin_norm", sentiment_col])
        loc = sub[sub["brand_origin_norm"] == "Local"][sentiment_col].astype(float)
        glo = sub[sub["brand_origin_norm"] == "Global"][sentiment_col].astype(float)

        if len(loc) < 20 or len(glo) < 20:
            # not enough sample for stable tests
            rows.append({
                **{group_cols[i]: keys[i] for i in range(len(group_cols))},
                "n_local": len(loc),
                "n_global": len(glo),
                "local_mean": loc.mean() if len(loc) else np.nan,
                "global_mean": glo.mean() if len(glo) else np.nan,
                "mean_diff_local_minus_global": (loc.mean() - glo.mean()) if (len(loc) and len(glo)) else np.nan,
                "welch_t_pvalue": np.nan,
                "levene_pvalue": np.nan,
                "note": "Insufficient sample (need >=20 per group)"
            })
            continue

        # Welch t-test for mean
        t_p = ttest_ind(loc, glo, equal_var=False).pvalue

        # Levene test for variance (center="median" is robust)
        l_p = levene(loc, glo, center="median").pvalue

        rows.append({
            **{group_cols[i]: keys[i] for i in range(len(group_cols))},
            "n_local": len(loc),
            "n_global": len(glo),
            "local_mean": loc.mean(),
            "global_mean": glo.mean(),
            "mean_diff_local_minus_global": loc.mean() - glo.mean(),
            "welch_t_pvalue": t_p,
            "local_var": loc.var(ddof=1),
            "global_var": glo.var(ddof=1),
            "var_ratio_local_over_global": loc.var(ddof=1) / glo.var(ddof=1) if glo.var(ddof=1) > 0 else np.nan,
            "levene_pvalue": l_p,
            "note": ""
        })

    out = pd.DataFrame(rows)
    out = out.sort_values(list(group_cols)).reset_index(drop=True)
    return out


def regression_origin_country_interaction(df: pd.DataFrame, sentiment_col: str) -> str:
    """
    OLS regression:
      sentiment ~ origin + platform + country + origin:country
    This tests moderation (interaction).
    """
    d = df.dropna(subset=["brand_origin_norm", "country", "source_platform", sentiment_col]).copy()
    d = d[d["brand_origin_norm"].isin(["Local", "Global"])].copy()

    # keep only countries with enough data to avoid super sparse interaction
    country_counts = d["country"].value_counts()
    keep_countries = country_counts[country_counts >= 100].index
    d = d[d["country"].isin(keep_countries)].copy()

    # OLS with categorical factors
    # brand_origin_norm treated as categorical; reference category chosen automatically (alphabetical).
    formula = f"{sentiment_col} ~ C(brand_origin_norm) + C(source_platform) + C(country) + C(brand_origin_norm):C(country)"
    model = smf.ols(formula, data=d).fit(cov_type="HC3")  # robust SEs

    return model.summary().as_text()


# -----------------------------
# Main
# -----------------------------
def main():
    # Step 1: write RQs + hypotheses
    write_rqs_and_hypotheses(RQ_OUT)

    # Load dataset with sentiment
    df = pd.read_csv(DATA_PATH)

    # Expecting a cleaned text column; fallback if your file uses a different name
    text_col = "text_clean" if "text_clean" in df.columns else ("text" if "text" in df.columns else None)
    if text_col is None:
        raise ValueError("No text column found. Expected text_clean or text.")

    # Normalize brand origin to Local/Global
    df = normalize_brand_origin(df)
    df = df.dropna(subset=["brand_origin_norm"]).copy()

    # Pick sentiment measure
    sentiment_col = pick_sentiment_column(df)
    print("Using sentiment column:", sentiment_col)

    # -------------------------
    # Step 2: LDA Topic Modeling
    # -------------------------
    # Run LDA on the entire corpus
    lda, vec, doc_topic, feat = run_lda(df, text_col=text_col, n_topics=8)

    topics_df = top_words_per_topic(lda, feat, top_n=14)
    topics_df.to_csv(LDA_TOPIC_OUT, index=False)

    # Attach dominant topic to each doc
    dominant_topic = doc_topic.argmax(axis=1)
    df_topics = df[["source_platform", "country", "brand_origin_norm", "retailer", text_col]].copy()
    df_topics["dominant_topic"] = dominant_topic

    # Save per-doc topic distribution (optional but useful)
    doc_topic_df = pd.DataFrame(doc_topic, columns=[f"topic_{i}" for i in range(doc_topic.shape[1])])
    doc_topic_df = pd.concat([df[["source_platform","country","brand_origin_norm","retailer"]].reset_index(drop=True),
                              doc_topic_df], axis=1)
    doc_topic_df.to_csv(LDA_DOC_TOPIC_OUT, index=False)

    # -------------------------
    # Step 3: Local vs Global Stats
    # -------------------------
    stats_df = stats_local_global(df, sentiment_col=sentiment_col, group_cols=("source_platform", "country"))
    stats_df.to_csv(STATS_OUT, index=False)

    reg_text = regression_origin_country_interaction(df, sentiment_col=sentiment_col)
    with open(REG_OUT, "w", encoding="utf-8") as f:
        f.write(reg_text)

    # Simple console outputs
    print("\nSaved outputs to:", OUT_DIR)
    print("- RQs/Hypotheses:", RQ_OUT)
    print("- LDA topics:", LDA_TOPIC_OUT)
    print("- LDA doc-topic matrix:", LDA_DOC_TOPIC_OUT)
    print("- Local vs Global tests:", STATS_OUT)
    print("- Regression summary:", REG_OUT)


if __name__ == "__main__":
    main()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation

# -----------------------------
# Load data
# -----------------------------
DATA_PATH = "combined_with_sentiment.csv"
df = pd.read_csv(DATA_PATH)

# -----------------------------
# Helpers
# -----------------------------
def pick_sentiment_column(df):
    for c in ["ft_bert_score", "bert_score", "vader_compound", "textblob_polarity"]:
        if c in df.columns:
            return c
    raise ValueError("No sentiment column found")

def normalize_brand_origin(df):
    s = df["brand_origin"].astype(str).str.lower()
    def f(x):
        if "local" in x or "domestic" in x:
            return "Local"
        if "global" in x or "foreign" in x or "international" in x:
            return "Global"
        return np.nan
    df["brand_origin_norm"] = s.apply(f)
    return df

# -----------------------------
# Prep
# -----------------------------
df = normalize_brand_origin(df)
df = df[df["brand_origin_norm"].isin(["Local", "Global"])].copy()

sent_col = pick_sentiment_column(df)
text_col = "text_clean" if "text_clean" in df.columns else "text"

# =============================
# FIGURE 1: Boxplot (Local vs Global)
# =============================
plt.figure()
plt.boxplot(
    [
        df[df["brand_origin_norm"] == "Local"][sent_col],
        df[df["brand_origin_norm"] == "Global"][sent_col]
    ],
    labels=["Local", "Global"],
    showfliers=False
)
plt.title("Sentiment Distribution: Local vs Global")
plt.ylabel("Sentiment score")
plt.show()

# =============================
# FIGURE 2: Histogram
# =============================
plt.figure()
plt.hist(df[df["brand_origin_norm"] == "Local"][sent_col], bins=30, alpha=0.6, label="Local")
plt.hist(df[df["brand_origin_norm"] == "Global"][sent_col], bins=30, alpha=0.6, label="Global")
plt.title("Sentiment Distribution (Histogram)")
plt.xlabel("Sentiment score")
plt.ylabel("Frequency")
plt.legend()
plt.show()

# =============================
# FIGURE 3: Mean sentiment by Country
# =============================
top_countries = df["country"].value_counts().head(10).index

by_country = (
    df[df["country"].isin(top_countries)]
    .groupby(["country", "brand_origin_norm"])[sent_col]
    .mean()
    .reset_index()
)

pivot = by_country.pivot(index="country", columns="brand_origin_norm", values=sent_col)
pivot.plot(kind="bar")
plt.title("Mean Sentiment by Country (Top 10): Local vs Global")
plt.ylabel("Mean sentiment")
plt.show()

# =============================
# FIGURE 4: Mean sentiment by Country × Platform
# =============================
for platform in df["source_platform"].dropna().unique():
    sub = df[(df["source_platform"] == platform) & (df["country"].isin(top_countries))]
    if sub.empty:
        continue

    by_cp = (
        sub.groupby(["country", "brand_origin_norm"])[sent_col]
           .mean()
           .reset_index()
    )

    piv = by_cp.pivot(index="country", columns="brand_origin_norm", values=sent_col)
    piv.plot(kind="bar")
    plt.title(f"Mean Sentiment by Country — {platform}")
    plt.ylabel("Mean sentiment")
    plt.show()

# =============================
# FIGURE 5: Variance by Country
# =============================
var_country = (
    df[df["country"].isin(top_countries)]
    .groupby(["country", "brand_origin_norm"])[sent_col]
    .var()
    .reset_index()
)

var_piv = var_country.pivot(index="country", columns="brand_origin_norm", values=sent_col)
var_piv.plot(kind="bar")
plt.title("Sentiment Variance by Country (Top 10)")
plt.ylabel("Variance")
plt.show()

# =============================
# LDA TOPIC MODELING
# =============================
vectorizer = CountVectorizer(
    stop_words="english",
    max_df=0.9,
    min_df=10,
    max_features=6000,
    ngram_range=(1, 2)
)

X = vectorizer.fit_transform(df[text_col].fillna("").astype(str))
lda = LatentDirichletAllocation(n_components=8, random_state=42)
doc_topic = lda.fit_transform(X)

# Show top words per topic (printed, not saved)
terms = vectorizer.get_feature_names_out()
print("\nTop words per topic:")
for i, topic in enumerate(lda.components_):
    top_words = [terms[j] for j in topic.argsort()[-10:][::-1]]
    print(f"Topic {i}: {', '.join(top_words)}")

df["dominant_topic"] = doc_topic.argmax(axis=1)

# =============================
# FIGURE 6: Topic prevalence by Platform & Origin
# =============================
topic_prev = (
    df.groupby(["source_platform", "brand_origin_norm", "dominant_topic"])
      .size()
      .reset_index(name="n")
)

topic_prev["prop"] = topic_prev["n"] / topic_prev.groupby(
    ["source_platform", "brand_origin_norm"]
)["n"].transform("sum")

for platform in topic_prev["source_platform"].unique():
    sub = topic_prev[topic_prev["source_platform"] == platform]
    piv = sub.pivot(index="brand_origin_norm", columns="dominant_topic", values="prop").fillna(0)
    piv.plot(kind="bar", stacked=True)
    plt.title(f"Topic Prevalence by Brand Origin — {platform}")
    plt.ylabel("Proportion")
    plt.show()

# =============================
# FIGURE 7: Mean sentiment per Topic
# =============================
topic_sent = (
    df.groupby(["source_platform", "dominant_topic"])[sent_col]
      .mean()
      .reset_index()
)

for platform in topic_sent["source_platform"].unique():
    sub = topic_sent[topic_sent["source_platform"] == platform]
    plt.figure()
    plt.bar(sub["dominant_topic"].astype(str), sub[sent_col])
    plt.title(f"Mean Sentiment by Topic — {platform}")
    plt.xlabel("Topic")
    plt.ylabel("Mean sentiment")
    plt.show()
```
